```python
Spell checking and correction: Apply minimum edit distance between two strings for spelling
correction.

What you did:
Compared two strings
Found insert, delete, replace operations
🎯 Why we do it:

To measure how similar two words are.

🌍 Use in NLP:
Spell checkers (Google, Grammarly)
Auto-correction systems
```

```python
# Process 1: Input Words
s1 = "kitten"
s2 = "sitting"
```

```python
# Process 2: MED + Operations (Simple Version)

# Function to calculate Minimum Edit Distance (MED) between two strings
def med(str1, str2):

    # Get lengths of both strings
    m = len(str1)
    n = len(str2)

    # Create a 2D matrix (m+1 x n+1) initialized with 0
    T = []
    for i in range(m+1):
        row = []
        for j in range(n+1):
            row.append(0)
        T.append(row)

    # Fill the matrix using dynamic programming
    for i in range(m+1):
        for j in range(n+1):

            # If both strings are empty
            if i == 0 and j == 0:
                T[i][j] = 0

            # If first string is empty → insert all characters of second string
            elif i == 0:
                T[i][j] = T[i][j-1] + 1

            # If second string is empty → delete all characters of first string
            elif j == 0:
                T[i][j] = T[i-1][j] + 1

            # If characters match → no operation needed
            elif str1[i-1] == str2[j-1]:
                T[i][j] = T[i-1][j-1]

            # If characters do not match → take minimum of insert, delete, replace
            else:
                T[i][j] = min(T[i][j-1], T[i-1][j], T[i-1][j-1]) + 1

    # Backtracking to find the sequence of operations
    i, j = m, n
    ops = []

    # Traverse the matrix from bottom-right to top-left
    while i > 0 or j > 0:

        # If characters match → no operation
        if i > 0 and j > 0 and str1[i-1] == str2[j-1]:
            ops.append("Match")
            i -= 1
            j -= 1

        # If replace operation was used
        elif i > 0 and j > 0 and T[i][j] == T[i-1][j-1] + 1:
            ops.append("Replace")
            i -= 1
            j -= 1

        # If delete operation was used
        elif i > 0 and T[i][j] == T[i-1][j] + 1:
            ops.append("Delete")
            i -= 1

        # Otherwise, insert operation
        else:
            ops.append("Insert")
            j -= 1

    # Reverse operations to get correct order
    ops.reverse()

    # Return minimum edit distance, operations list, and matrix
    return T[m][n], ops, T
```

```python
# Process 3: Show Matrix

# Call the med() function to get distance, operations, and matrix
dist, ops, matrix = med(s1, s2)

# Loop through each row of the matrix
for r in matrix:
    # Print each row (shows the edit distance table)
    print(r)
```

```python
# Process 4: Result

# Print the final Minimum Edit Distance value
print("MED =", dist)

# Count and print number of replace operations
print("Replace:", ops.count("Replace"))

# Count and print number of insert operations
print("Insert :", ops.count("Insert"))

# Count and print number of delete operations
print("Delete :", ops.count("Delete"))
```

```python
1. What is Minimum Edit Distance (MED)?

It is the minimum number of operations (insert, delete, replace) needed to convert one string into another.

2. What operations are used in MED?

Insertion, deletion, and substitution (replacement).

3. How is MED useful in spell checking?

It finds the closest correct word by comparing a misspelled word with dictionary words.

4. What is the time complexity of MED?

O(m × n), where m and n are lengths of the two strings.

5. Why do we use a matrix in MED?

To store intermediate results and avoid recomputation using dynamic programming.

6. What does each cell in the matrix represent?

It represents the edit distance between substrings of the two words.

7. Why do we initialize the first row and column?

To handle cases where one string is empty (all insertions or deletions).

8. What happens when characters match?

No operation is needed, so we copy the diagonal value.

9. What happens when characters do not match?

We take the minimum of insert, delete, and replace operations + 1.

10. What is backtracking in MED?

It is used to find the sequence of operations that leads to the minimum edit distance.

🔸 Deep / Conceptual QnA
11. Why is MED preferred for spell correction?

Because it measures similarity between words and finds the closest match efficiently.

12. Can MED handle real-world spelling errors perfectly?

No, because it does not consider context or word frequency.

13. What are the limitations of MED?

It ignores context, pronunciation, and semantic meaning.

14. How can MED be improved for better spell checking?

By combining it with dictionaries, frequency models, or language models.

15. What is weighted edit distance?

Different operations have different costs (e.g., replace cost ≠ insert cost).

16. Why might replacement cost be lower than insertion?

Because typing errors often involve wrong letters rather than missing ones.

17. How would you handle large dictionaries efficiently?

Use indexing, tries, or limit comparisons using heuristics.

18. How does MED differ from Hamming Distance?

MED allows insert/delete; Hamming only works for equal-length strings.

19. How would you use MED in autocorrect systems?

Compare input word with dictionary words and select the one with minimum distance.

20. Can MED be used in speech recognition?

Yes, to match predicted words with closest valid words.
```
