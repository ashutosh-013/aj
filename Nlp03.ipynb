```python
3. Viterbi Decoding Algorithm for POS: Given a sequence of words, implement the Viterbi algorithm
to find the most likely sequence of POS tags by estimating emission and transition probabilities from a
tagged corpus.

    ✔ What you did:
Assigned POS tags using probabilities (HMM + Viterbi)
🎯 Why we do it:

To find most likely grammatical structure of a sentence.

🌍 Use in NLP:
Grammar analysis
Speech recognition
Information extraction
```

```python
# Input sentence (words to be tagged)
sentence = ["I", "love", "coding"]

# Possible POS tags (states in HMM)
tags = ["PRON", "VERB", "NOUN"]

# Transition probabilities:
# Probability of moving from one tag to another
transition_prob = {
    "START": {"PRON": 0.6, "VERB": 0.2, "NOUN": 0.2},  # Start state probabilities
    "PRON": {"VERB": 0.7, "NOUN": 0.2, "PRON": 0.1},   # From PRON to others
    "VERB": {"NOUN": 0.6, "VERB": 0.3, "PRON": 0.1},   # From VERB to others
    "NOUN": {"VERB": 0.4, "NOUN": 0.4, "PRON": 0.2}    # From NOUN to others
}

# Emission probabilities:
# Probability of a word being generated from a tag
emission_prob = {
    "PRON": {"I": 0.9, "love": 0.1, "coding": 0.0},    # PRON emits words
    "VERB": {"I": 0.0, "love": 0.8, "coding": 0.2},    # VERB emits words
    "NOUN": {"I": 0.0, "love": 0.1, "coding": 0.9}     # NOUN emits words
}
```

```python
# Initialize Viterbi matrix (stores probabilities at each step)
V = [{}]

# Dictionary to store the best path for each tag
path = {}

# ------------------ Initialization ------------------
# Calculate probabilities for the first word in the sentence
for tag in tags:
    # Probability = start transition * emission of first word
    V[0][tag] = transition_prob["START"][tag] * emission_prob[tag][sentence[0]]
    
    # Initialize path with current tag
    path[tag] = [tag]

# ------------------ Recursion ------------------
# Loop through remaining words in the sentence
for t in range(1, len(sentence)):
    V.append({})        # Add new column for time step
    new_path = {}       # Store updated paths

    for curr in tags:   # Current tag
        max_prob = -1
        best_prev = None

        # Check all possible previous tags
        for prev in tags:
            # Compute probability of transition + emission
            prob = V[t-1][prev] * transition_prob[prev][curr] * emission_prob[curr][sentence[t]]
            
            # Select the maximum probability path
            if prob > max_prob:
                max_prob = prob
                best_prev = prev

        # Store best probability for current tag
        V[t][curr] = max_prob
        
        # Update path with best previous tag
        new_path[curr] = path[best_prev] + [curr]

    # Update paths for next iteration
    path = new_path

# ------------------ Termination ------------------
# Find the tag with highest probability at the last word
best_tag = max(V[-1], key=V[-1].get)

# Get the best sequence of tags
best_sequence = path[best_tag]
```

```python
# Print the input sentence
print("Sentence:", sentence)

# Print the predicted POS tag sequence (best path from Viterbi)
print("POS Tags:", best_sequence)
```

```python
11. Why dynamic programming is used in Viterbi?

To avoid recomputation and improve efficiency.

12. What is the time complexity of Viterbi?

O(n × k²), where n = sentence length, k = number of tags.

13. What is the problem with zero probabilities?

If any probability is zero, the path becomes impossible.

14. How to handle zero probabilities?

Use smoothing techniques like Laplace smoothing.

15. Why do we multiply probabilities?

Because we assume independence and use joint probability.

16. What is the independence assumption in HMM?

Each state depends only on the previous state, not the entire sequence.

17. What is the Markov assumption?

Current state depends only on the previous state.

18. What is the difference between greedy and Viterbi?

Greedy chooses local best; Viterbi finds global best sequence.

19. Why use log probabilities in real systems?

To avoid underflow when multiplying very small numbers.

20. How are emission and transition probabilities obtained?

From a tagged corpus using frequency counts.

🔥 High-Impact Viva Questions
21. Why is Viterbi better than brute force?

Because brute force checks all sequences (exponential), while Viterbi is efficient (polynomial).

22. What are limitations of Viterbi?

Depends on probability accuracy, ignores long-range context.

23. How would you improve POS tagging accuracy?

Use better training data, smoothing, or deep learning models.

24. Can Viterbi be used outside NLP?

Yes, in speech recognition, bioinformatics, and decoding problems.

25. Why HMM is considered a generative model?

Because it models how data (words) is generated from hidden states.
```
