```python
# Step 1: Import

# Import NLTK library for NLP tasks
import nltk

# Import tokenizers to split text into words and sentences
from nltk.tokenize import word_tokenize, sent_tokenize

# Import POS tagger (assigns part-of-speech tags to words)
from nltk import pos_tag

# Download 'punkt' dataset for tokenization
nltk.download('punkt')

# Download pretrained POS tagger model
nltk.download('averaged_perceptron_tagger')
```

```python
# Step 2: Input Text
text = """John went to the market. He bought apples.
Mary saw John there. She greeted him.
The dog was barking loudly. It was hungry."""
```

```python
# Step 3: Sentence Split

# Split the input text into individual sentences
sentences = sent_tokenize(text)

# Display the list of sentences
sentences
```

```python
# Step 4: Name lists
male = ['john', 'ram', 'rohit']
female = ['mary', 'sita', 'priya']
```

```python
# Step 5: Extract nouns

# Lists to store proper nouns (persons) and common nouns (objects)
persons, objects = [], []

# Loop through each sentence
for s in sentences:
    
    # Tokenize sentence into words and assign POS tags
    for w, t in pos_tag(word_tokenize(s)):
        
        # Convert word to lowercase for consistency
        w = w.lower()
        
        # If tag is proper noun (NNP) → likely a person's name
        if t == "NNP":
            persons.append(w)
        
        # If tag is common noun (NN) → likely an object
        elif t == "NN":
            objects.append(w)
```

```python
# Step 6: Co-reference Resolution

# Loop through each sentence
for s in sentences:
    
    # Tokenize and POS tag each word
    for w, t in pos_tag(word_tokenize(s)):
        
        # Convert word to lowercase
        w = w.lower()

        # Check if word is a pronoun (PRP)
        if t == "PRP":
            
            # For male pronouns → map to last matching male person
            if w in ["he", "him"]:
                print(w, "→", next((x for x in reversed(persons) if x in male), ""))

            # For female pronouns → map to last matching female person
            elif w in ["she", "her"]:
                print(w, "→", next((x for x in reversed(persons) if x in female), ""))

            # For neutral pronoun → map to last object
            elif w == "it":
                print(w, "→", objects[-1] if objects else "")
```
