```python
# Import NLTK library
import nltk

# Import WordNet (lexical database for meanings of words)
from nltk.corpus import wordnet

# Import Lesk algorithm for Word Sense Disambiguation (WSD)
from nltk.wsd import lesk
```

```python
# Download WordNet dataset (used for word meanings and synonyms)
nltk.download('wordnet')

# Download multilingual WordNet data (supports extended word meanings)
nltk.download('omw-1.4')
```

```python
s1 = "I went to the bank to deposit money"
s2 = "He sat on the bank of the river"
```

```python
# Split sentence s1 into individual words (tokens)
words1 = s1.split()

# Split sentence s2 into individual words (tokens)
words2 = s2.split()
```

```python
# Lesk senses

# Apply Lesk algorithm on first sentence to find meaning of "bank"
sense1 = lesk(words1, "bank")

# Apply Lesk algorithm on second sentence to find meaning of "bank"
sense2 = lesk(words2, "bank")
```

```python
# Function to calculate overlap score between context and sense

def overlap_score(context, sense):
    
    # Get words from the sense definition and convert to lowercase set
    sense_words = set(sense.definition().lower().split())
    
    # Convert context words into a set
    context_words = set(context)
    
    # Return number of common words between context and sense definition
    return len(sense_words.intersection(context_words))
```

```python
# Compute scores

# Calculate overlap score for first sentence and its sense
score1 = overlap_score(words1, sense1)

# Calculate overlap score for second sentence and its sense
score2 = overlap_score(words2, sense2)
```

```python
# OUTPUT
print("Sentence 1 Sense:", sense1)
print("Meaning:", sense1.definition())
print("Overlap Score:", score1)
```

```python
print("\nSentence 2 Sense:", sense2)
print("Meaning:", sense2.definition())
print("Overlap Score:", score2)
```
