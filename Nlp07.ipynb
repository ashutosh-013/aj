```python
# Step 1: Load Dataset

# Import json module to read JSON data
import json

# Create an empty list to store dataset
data = []

# Open the dataset file in read mode with UTF-8 encoding
with open("Cell_Phones_and_Accessories_5.json", "r", encoding="utf-8") as f:
    
    # Read file line by line (each line is a JSON object)
    for line in f:
        
        # Convert JSON string to Python dictionary and add to list
        data.append(json.loads(line))

# Print total number of records in dataset
print("Dataset size:", len(data))
```

```python
# Step 3: Preprocessing

# Import required libraries
import nltk, re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download tokenizer and stopwords datasets
nltk.download('punkt')
nltk.download('stopwords')

# Load English stopwords into a set for fast lookup
stop_words = set(stopwords.words('english'))

# Function to clean and tokenize text
def preprocess(text):
    
    # Convert to lowercase and remove non-alphabet characters
    text = re.sub(r'[^a-z\s]', '', text.lower())
    
    # Split text into words
    words = word_tokenize(text)
    
    # Remove stopwords and return cleaned tokens
    return [w for w in words if w not in stop_words]

# Apply preprocessing to each sentence (skip empty ones)
tokenized_sentences = [preprocess(s) for s in sentences if s]
```

```python
# Step 4: Train Word2Vec

# Import Word2Vec model from gensim
from gensim.models import Word2Vec

# Train Word2Vec model on tokenized sentences
w2v_model = Word2Vec(
    
    # Input data: list of tokenized sentences
    sentences=tokenized_sentences,
    
    # Size of word vectors (each word → 100-dimensional vector)
    vector_size=100,
    
    # Context window size (number of nearby words considered)
    window=5,
    
    # Ignore words that appear less than 5 times
    min_count=5
)
```

```python
# Similarity

# Calculate and print similarity between "phone" and "mobile"
# Returns a value between -1 and 1 (higher = more similar)
print("Similarity (phone, mobile):",
      w2v_model.wv.similarity("phone", "mobile"))
```

```python
# Step 5: Train FastText

# Import FastText model from gensim
from gensim.models import FastText

# Train FastText model on tokenized sentences
ft_model = FastText(
    
    # Input data: list of tokenized sentences
    sentences=tokenized_sentences,
    
    # Size of word vectors (each word → 100-dimensional vector)
    vector_size=100,
    
    # Context window size
    window=5,
    
    # Ignore words that appear less than 5 times
    min_count=5
)

# Print heading for FastText results
print("\nFastText Similar Words:")

# Find and print words similar to "phone"
print(ft_model.wv.most_similar("phone"))

# Calculate and print similarity between "phone" and "mobile"
print("Similarity (phone, mobile):",
      ft_model.wv.similarity("phone", "mobile"))
```

```python
1. What are word embeddings?

Word embeddings are numerical vector representations of words that capture meaning and relationships.

2. Why do we need word embeddings?

Because machines cannot understand text directly, so words are converted into vectors.

3. What is Word2Vec?

Word2Vec is a model that learns word meanings based on surrounding context.

4. What is FastText?

FastText is an extension of Word2Vec that uses subword (character-level) information.

5. What is GloVe?

GloVe is a word embedding model based on global word co-occurrence statistics.

6. What is similarity in word embeddings?

It measures how close two word vectors are in meaning.

7. What does most_similar() do?

It returns words closest in meaning to a given word.

8. Why do we use a window in Word2Vec?

It defines how many surrounding words are considered for context.

9. What is min_count?

It ignores words that appear less than a specified number of times.

10. What is vector size?

It defines the number of dimensions in word representation.

🔸 Deep / Conceptual QnA
11. How do word embeddings capture meaning?

Words used in similar contexts have similar vectors (distributional hypothesis).

12. Difference between Word2Vec and FastText?

Word2Vec treats words as whole units; FastText uses subword information.

13. Why is FastText better for rare words?

Because it breaks words into character n-grams.

14. Difference between GloVe and Word2Vec?

Word2Vec is prediction-based; GloVe is count-based.

15. What is semantic similarity?

Words having similar meanings (e.g., phone and mobile).

16. What is syntactic similarity?

Words with similar grammatical roles (e.g., run and walk).

17. What is the distributional hypothesis?

Words appearing in similar contexts have similar meanings.

18. Why are embeddings better than one-hot encoding?

Embeddings capture meaning; one-hot vectors are sparse and meaningless.

19. What is the advantage of dense vectors?

They reduce dimensionality and capture relationships.

20. What is cosine similarity?

A measure used to compare similarity between word vectors.

🔥 High-Impact Viva Questions
21. Why is FastText better for noisy data?

Because it handles misspellings using subword information.

22. What happens if a word is not in vocabulary?

Word2Vec fails, but FastText can still generate a vector.

23. Where are word embeddings used?

Chatbots, search engines, translation, sentiment analysis.

24. What is the limitation of Word2Vec?

It cannot handle out-of-vocabulary words.

25. How do embeddings improve NLP models?

They provide semantic understanding instead of raw text.
```
