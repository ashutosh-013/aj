```python
import pandas as pd

df = pd.read_csv("winequality_main.csv")
df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
# Step: Separate features from target variable

# Remove the target column "quality" from the dataset
# X contains all input features used for prediction
X = df.drop("quality", axis=1)
```

```python
# Import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Create scaler object (standardizes data: mean=0, std=1)
scaler = StandardScaler()

# Fit scaler on data and transform features into scaled form
X_scaled = scaler.fit_transform(X)
```

```python
# Import functions for hierarchical clustering and visualization
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Perform hierarchical clustering using Ward’s method
linked = linkage(X_scaled, method='ward')

# Set figure size for better visualization
plt.figure(figsize=(10,5))

# Plot dendrogram to show cluster hierarchy
dendrogram(linked)

# Add title to the plot
plt.title("Dendrogram - Wine Dataset")
```

```python
# Perform hierarchical clustering using Complete linkage method
linked = linkage(X_scaled, method='complete')

# Set figure size for better visualization
plt.figure(figsize=(12,6))

# Plot dendrogram (show only last 30 clusters for simplicity)
dendrogram(linked, truncate_mode='lastp', p=30)

# Add title to the plot
plt.title("Dendrogram - Wine Dataset")
```

```python
# Import Agglomerative Clustering algorithm
from sklearn.cluster import AgglomerativeClustering

# Create model with 3 clusters
model = AgglomerativeClustering(n_clusters=3)

# Fit model and predict cluster labels for each data point
labels = model.fit_predict(X_scaled)

# Print unique cluster labels formed
print(set(labels))
```

```python
plt.figure()

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels)

plt.title("Wine Clustering")
plt.show()
```

```python
print("Number of clusters:", len(set(labels)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=labels)
plt.title("Cluster Distribution")
```

```python
1. What is clustering?

Clustering is an unsupervised learning technique that groups similar data points together.

2. What is Agglomerative Clustering?

It is a bottom-up hierarchical clustering method where each point starts as its own cluster and merges step by step.

3. Why is it called hierarchical clustering?

Because it forms a tree-like structure (dendrogram) of clusters.

4. What is a dendrogram?

A tree diagram that shows how clusters are merged at different distances.

5. What dataset is used here?

Wine Quality dataset from UCI Machine Learning Repository.

6. Why do we scale data before clustering?

Because clustering is distance-based, and scaling ensures all features contribute equally.

7. What is linkage method?

It defines how distance between clusters is calculated (single, complete, ward, average).

8. What is Ward’s method?

It minimizes variance within clusters while merging.

9. Why use complete linkage?

It considers the maximum distance between points in clusters.

10. What does n_clusters=3 mean?

It means the algorithm will form 3 groups of data.

🔸 Deep / Conceptual QnA
11. Is clustering supervised or unsupervised?

Unsupervised, because there are no labels used during training.

12. How is clustering different from classification?

Classification uses labeled data, clustering does not.

13. Why is hierarchical clustering useful?

It does not require pre-defining number of clusters initially.

14. What is the main drawback of agglomerative clustering?

It is computationally expensive for large datasets.

15. How do you decide number of clusters from dendrogram?

By cutting the dendrogram at the longest vertical distance.

16. What is Euclidean distance?

It measures straight-line distance between two points.

17. Why is scaling important in wine dataset?

Because features like alcohol and acidity have different ranges.

18. Can clustering predict wine quality exactly?

No, it groups similar wines but does not guarantee exact quality labels.

19. What is the difference between Ward and Complete linkage?

Ward minimizes variance, Complete uses maximum distance between clusters.

20. What are applications of clustering?

Customer segmentation, anomaly detection, image grouping, and recommendation systems.

🔥 High-Impact Viva Questions
21. Why is clustering used for wine quality?

To group similar wine samples based on chemical properties.

22. What happens if we don’t scale data?

Features with large values dominate clustering results.

23. Why is dendrogram useful?

It helps visualize hierarchical relationships and choose cluster count.

24. Why is agglomerative clustering called greedy?

Because it merges closest clusters step by step without backtracking.

25. Where is hierarchical clustering used in real life?

Genetics, document clustering, and market segmentation.
```
