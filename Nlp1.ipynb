```python
 Text Preprocessing - Implement text preprocessing techniques, including tokenization and
normalization, and apply it to any text data.
✔ What you did:
Split text into words/sentences
Removed punctuation, stopwords
Applied stemming & lemmatization
🎯 Why we do it:

To clean raw text and convert it into a machine-readable form.

🌍 Use in NLP:
Chatbots
Search engines
Sentiment analysis
```

```python
# Step 1: Input Text

# Store a paragraph about Artificial Intelligence in a variable named 'para'
para = "Artificial Intelligence is the simulation of human intelligence by computer systems, enabling machines to learn, reason, solve problems, and perceive their environment. It utilizes algorithms and data to perform tasks like language translation, image recognition, and decision-making." 

# Print the content of the variable 'para' to display the text
print(para)
```

```python
# Step 2: Tokenization

# Import required libraries
import nltk, string

# Import functions to split text into words and sentences
from nltk.tokenize import word_tokenize, sent_tokenize

# Download the 'punkt' dataset (used for tokenizing text)
nltk.download('punkt')

# Convert the paragraph into individual words (tokens)
words = word_tokenize(para)

# Print the list of word tokens
print(words)

# Convert the paragraph into sentences
# and print the list of sentences
print(sent_tokenize(para))
```

```python
# Step 3–5: Lowercase + Remove Punctuation + Remove Stopwords

# Import stopwords (common words like 'is', 'the', etc.)
from nltk.corpus import stopwords

# Download the stopwords dataset
nltk.download('stopwords')

# Convert all words to lowercase and keep only alphabetic words (remove punctuation/numbers)
words = [w.lower() for w in words if w.isalpha()]

# Remove common stopwords from the word list
words = [w for w in words if w not in stopwords.words('english')]

# Print the cleaned list of words
print(words)
```

```python
# Step 6: Stemming (with arrows)

# Import Porter Stemmer (used to reduce words to root form)
from nltk.stem import PorterStemmer

# Create an object of the stemmer
ps = PorterStemmer()

# Loop through each word in the list
for w in words:
    # Print original word and its stemmed (root) form
    print(w, "->", ps.stem(w))
```

```python
# Step 7: Final Output

# Apply stemming to all words, join them into a single string, and print the result
print("Stemmed:", " ".join([ps.stem(w) for w in words]))
```

```python
# Step 8: Lemmatization (with arrows)

# Import WordNet Lemmatizer (used to get base/dictionary form of words)
from nltk.stem import WordNetLemmatizer

# Download the WordNet dataset (required for lemmatization)
nltk.download('wordnet')

# Create a lemmatizer object
lemma = WordNetLemmatizer()

# Loop through each word in the list
for w in words:
    # Print original word and its lemmatized (base) form
    print(w, "->", lemma.lemmatize(w))
```

```python
# Step 9: Final Output

# Apply lemmatization to all words, join them into a single string, and print the result
print("Lemmatized:", " ".join([lemma.lemmatize(w) for w in words]))
```

```python
1. What is text preprocessing and why is it important?
It is cleaning and transforming raw text into a usable format. It improves model accuracy by removing noise.

2. Tokenization vs Normalization
Tokenization = splitting text into words/sentences.
Normalization = cleaning and standardizing text (lowercase, remove noise).

3. Why convert to lowercase?
To treat words like “AI” and “ai” as the same, reducing duplicates.

4. Why remove punctuation?
It usually doesn’t add meaning and creates unnecessary tokens.

5. What are stopwords?
Common words like “the”, “is” with little meaning; removed to reduce noise.

6. Can removing stopwords be harmful?
Yes. Example: “not good” → removing “not” changes meaning.

7. Word vs Sentence tokenization
Word = splits into words.
Sentence = splits into sentences.

8. What does isalpha() do?
It keeps only alphabetic words and removes numbers/symbols.

9. What is stemming?
Cuts words to root form using rules (fast but less accurate).

10. What is lemmatization?
Converts words to dictionary form using context (more accurate).

11. Why does stemming give incorrect words?
Because it uses rules, not context (e.g., “studies” → “studi”).

12. Which is more accurate?
Lemmatization, because it uses vocabulary and context.

13. Role of 'punkt' and 'wordnet'?
'punkt' → tokenization
'wordnet' → lemmatization

14. Why use nltk.download()?
To download required datasets that NLTK functions depend on.

🔸 Conceptual Answers

15. Why is normalization important in sentiment analysis?
It ensures consistent text, helping models correctly detect sentiment.

16. When would you NOT remove punctuation?
In sentiment/emotion tasks (e.g., “!” shows excitement).

17. Handling numbers, emojis, URLs
Numbers → remove or convert
Emojis → convert to meaning (😊 → happy)
URLs → remove

18. Can preprocessing cause bias or loss?
Yes, removing words may lose important meaning or context.

19. Trade-off: cleaning vs information loss
More cleaning → less noise but possible loss of meaning.

20. Other languages preprocessing?
Needs language-specific rules, tokenizers, and stopwords.

21. Why not just split by spaces?
Because punctuation, contractions, and languages make it complex.

22. Deep learning vs traditional ML preprocessing
Traditional ML → heavy preprocessing
Deep learning → minimal preprocessing (models learn patterns)
```
